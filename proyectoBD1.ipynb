## Top 50 Fast-Food Chains in USA

En este conjunto de datos, tenemos información diversa sobre las 50 principales cadenas de comida rápida en los EE. UU. (2021). Las características clave de este conjunto de datos son: Cadenas de comida rápida, Ventas en todo el sistema de EE. UU. (Millones - dólares estadounidenses), Ventas promedio por unidad (Miles - dólares estadounidenses), Tiendas franquiciadas, Tiendas de la empresa, Unidades totales de 2021, Cambio total en unidades desde 2020 Nota: Algunos valores son estimados por QSR.

```
# Leer la tabla 'nombre_de_la_tabla'
df = spark.read.table('tabla_fixed')
# Verificar el resultado
display(df)
```

```
df = df.withColumnRenamed('U.S. Systemwide Sales (Millions - U.S Dollars)', 'ventas') \
       .withColumnRenamed('Average Sales per Unit (Thousands - U.S Dollars)', 'promedio_ventas_porunidad') \
       .withColumnRenamed('Franchised Stores', 'franquicicas') \
       .withColumnRenamed('Company Stores', 'company_stores') \
       .withColumnRenamed('numero_cambios', 'numero_cambios2020') \
       .withColumnRenamed('2021 Total Units', 'Total2021_Unidades')\
       .withColumnRenamed('Fast-Food Chains', 'cadenas')


# Verificar el resultado
display(df)
```

```
df.selectExpr("mean(ventas)", "median(ventas)", "approx_percentile(ventas, 0.25)", "approx_percentile(ventas, 0.75)").show()

# Medidas de dispersión
df.selectExpr("stddev(ventas)", "variance(ventas)").show()

# Medidas de posición
df.selectExpr("min(ventas)", "max(ventas)").show()
```

```
df.selectExpr("mean(promedio_ventas_porunidad)", "median(promedio_ventas_porunidad)", "approx_percentile(promedio_ventas_porunidad, 0.25)", "approx_percentile(ventas, 0.75)").show()

# Medidas de dispersión
df.selectExpr("stddev(promedio_ventas_porunidad)", "variance(promedio_ventas_porunidad)").show()

# Medidas de posición
df.selectExpr("min(promedio_ventas_porunidad)", "max(promedio_ventas_porunidad)").show()
```

```
df.selectExpr("mean(franquicicas)", "median(franquicicas)", "approx_percentile(franquicicas, 0.25)", "approx_percentile(franquicicas, 0.75)").show()

# Medidas de dispersión
df.selectExpr("stddev(franquicicas)", "variance(franquicicas)").show()

# Medidas de posición
df.selectExpr("min(franquicicas)", "max(franquicicas)").show()
```

```
df.selectExpr("mean(company_stores)", "median(company_stores)", "approx_percentile(company_stores, 0.25)", "approx_percentile(company_stores, 0.75)").show()

# Medidas de dispersión
df.selectExpr("stddev(company_stores)", "variance(company_stores)").show()

# Medidas de posición
df.selectExpr("min(company_stores)", "max(company_stores)").show()
```

```
df.selectExpr("mean(Total2021_Unidades)", "median(Total2021_Unidades)", "approx_percentile(Total2021_Unidades, 0.25)", "approx_percentile(Total2021_Unidades, 0.75)").show()

# Medidas de dispersión
df.selectExpr("stddev(Total2021_Unidades)", "variance(Total2021_Unidades)").show()

# Medidas de posición
df.selectExpr("min(Total2021_Unidades)", "max(Total2021_Unidades)").show()
```
